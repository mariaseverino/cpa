```python
!pip install subsetsum
import numpy as np
import subsetsum
import time
import matplotlib.pyplot as plt
```

```python
def gera_arranjos_aleatorio(n_elementos):
  A=np.random.randint(-100, 100, (n_elementos))
  return A
```

```python
def temSolucao(nums, target):
  inicio = time.time()

  has_solution = subsetsum.has_solution(nums, target)
  print(f"O conjunto: {nums} onde a soma é: {target} tem solução? {has_solution}")
  
  fim = time.time()
  tempo = fim - inicio

  return tempo
```

```python
def solucoes(nums, target):
  inicio = time.time()

  for solution in subsetsum.solutions(nums, target):
    subset = [nums[i] for i in solution]
    sum_value=sum(subset)
    print(f"{subset} se somados, reulta em: {sum_value}")

  fim = time.time()
  tempo = fim - inicio

  return tempo
```

```python
tempos = []
tempos2 = []

for i in range(5, 25):

  target = 0
  # nums = [-2, -1, 0, 1, 2]
  nums=gera_arranjos_aleatorio(i)

  tempo1 = temSolucao(nums, target)
  tempo2 = solucoes(nums, target)

  tempos.append(tempo1)
  tempos2.append(tempo2)
```

```python
print(tempos)
```

```python
plt.plot(tempos, label='Tem solução?')
plt.legend();
```

Observando o grafico percebemos que para verificar se há algum subconjunto presente no arranjo, o algoritmo varia muito, o que deve ser cuasando por alguns arranjos terem que fazer muitas verificações em subarranjos para saber se existe alguma solução.

```python
plt.plot(tempos2, label='Soluções')
plt.legend();
```

Observando o grafico percebemos que para verificar todos os subconjuntos presentes no arranjo, o algoritmo apresenta um comportamento exponencial.
